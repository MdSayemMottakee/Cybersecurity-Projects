#Task 2
Find the decrypted plaintext from the following:

n = 403212583809059293300199955599

c = 360112796689162206303023399384

e = 65537

Hint: Sympy is a great tool for complex mathematical operations

```python
from Crypto.Util.number import inverse
from sympy import factorint

n = 403212583809059293300199955599
e = 65537
c = 360112796689162206303023399384

factors = factorint(n)
if len(factors) != 2:
    raise ValueError("n does not have exactly two prime factors")

p, q = factors.keys()

phi_n = (p - 1) * (q - 1)

d = inverse(e, phi_n)

m = pow(c, d, n)

plaintext = bytes.fromhex(hex(m)[2:]).decode()

print("Decrypted plaintext:", plaintext)
```

#Task 3

Image: https://drive.google.com/file/d/1ea2M_L4RsJLUwGnj8LBaErxiek3CbAox/view?usp=sharing

Message: **birds_of_a_same_feather_flock_together**

Hide the message inside the green layer of the image and extract the message again to verify

```python
# Your code here
from PIL import Image

def hide_message(image_path, message):
    img = Image.open(image_path)

    binary_message = ''.join(format(ord(c), '08b') for c in message)
    binary_message += '00000000'

    width, height = img.size
    max_capacity = width * height

    if len(binary_message) > max_capacity:
        raise ValueError("Message too long to fit in image.")

    new_img = Image.new(img.mode, img.size)
    message_index = 0

    for y in range(height):
        for x in range(width):
            pixel = img.getpixel((x, y))
            r, g, b = pixel

            if message_index < len(binary_message):
                g_bin = format(g, '08b')
                g_bin = g_bin[:-1] + binary_message[message_index]
                g = int(g_bin, 2)
                message_index += 1

            new_img.putpixel((x, y), (r, g, b))

    new_img.save("encrypted_bird.png")
    print(" Message hidden inside 'encrypted_bird.png'")
```

```python
image_path = "/content/Evening_Clouds_Sky.jpeg"
message = "birds_of_a_same_feather_flock_together"
hide_message(image_path,message)
```

```python
def extract_message(image_path):
    img = Image.open(image_path)
    width, height = img.size

    binary_message = ''
    for y in range(height):
        for x in range(width):
            pixel = img.getpixel((x, y))
            g = pixel[1]
            g_bin = format(g, '08b')
            binary_message += g_bin[-1]

    message = ''
    for i in range(0, len(binary_message), 8):
        char = chr(int(binary_message[i:i+8], 2))
        if char == '\x00':
            break
        message += char
    return message
```
